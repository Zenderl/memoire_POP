Q1
évolution des adhésions entre 1957-1976 (prendre toutes les adhésions) cette requête permet de voir le nombre d'adhésion par genre sur une période de 15ans. séparé par période de 4ans sur la base des législatures cantonales pour regrouper les individus.

```python
import sqlite3
import pandas 
import matplotlib.pyplot as plt
import seaborn as sns
connection = sqlite3.connect('C:/Users/LZender/Documents/Travail/unine/histoire/mémoire/AEN/grand_conseil/memoire_POP/liste_grand_conseil.db')


cursor = connection.cursor()
```

# 1. évolution des adhésions

```python

result = cursor.execute("""
SELECT COUNT(pk_personne)
FROM personne
WHERE adhesion BETWEEN 1957 AND 1976
""")

print(cursor.fetchall()[0][0])
```

```python
annees = list(range(1957, 1978, 4))

periodes = []

for index, annee_item in enumerate(annees):
    if index + 1 < len(annees):
        periodes.append({
            'start': annees[index],
            'end': annees[index + 1] - 1
        })

for periode_item in periodes:
    start = periode_item['start']
    end = periode_item['end']
    cursor.execute("""
SELECT COUNT(pk_personne)
FROM personne
WHERE genre = 'F' AND adhesion BETWEEN ? AND ?
""", [start, end])
    result_femmes = cursor.fetchall()[0][0]
    cursor.execute("""
SELECT COUNT(pk_personne)
FROM personne
WHERE genre = 'M' AND adhesion BETWEEN ? AND ?
""", [start, end])
    result_hommes = cursor.fetchall()[0][0]
    periode_item['femmes'] = result_femmes
    periode_item['hommes'] = result_hommes

print(periodes)
```

```python
df_adhesion = pandas.DataFrame.from_dict(periodes)

df_adhesion['total'] = df_adhesion['femmes'] + df_adhesion['hommes']

df_adhesion['pourcentage_femmes'] = df_adhesion['femmes'] / df_adhesion['total'] * 100.00
df_adhesion['pourcentage_hommes'] = df_adhesion['hommes'] / df_adhesion['total'] * 100.00

df_adhesion.head()
```

```python
df_copy_adhesion = df_adhesion.drop(columns=['femmes', 'hommes', 'total'])

df_copy_adhesion.to_csv('CSV/question_01-evolution_adhesions.csv')

df_copy_adhesion.head()
```

## 1.2 répartition par périodes en graphe

```python
df_modif_adhesion = df_copy_adhesion.copy()
df_modif_adhesion['periodes'] = df_modif_adhesion['start'].astype(str) + ' ' + df_modif_adhesion['end'].astype(str)
df_modif_adhesion = df_modif_adhesion.drop(columns=['start', 'end'])


df_modif_adhesion.head()
```

```python
df_long_adhesion = df_modif_adhesion.melt(
    id_vars='periodes',
    value_vars=['pourcentage_femmes','pourcentage_hommes'],
    var_name='genre',
    value_name='pourcentage'
)

df_long_adhesion.head()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df_long_adhesion, x='periodes', y='pourcentage', hue='genre', palette='pastel')

# Rotation des labels x
plt.xticks(rotation=45, ha='right')

# Titre et axes
plt.title("Répartition des adhésions par genre  (en % sur l'ensemble)")
plt.ylabel("Pourcentage")
plt.xlabel("Période")

# Ajouter les valeurs sur les barres
for p in ax.patches:
    hauteur = p.get_height()
    if not pandas.isna(hauteur):  # au cas où il y a des NaN
        ax.annotate(f'{hauteur:.1f}%', 
                    (p.get_x() + p.get_width() / 2., hauteur),
                    ha='center', va='bottom', fontsize=9)

ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Ajustement des marges
plt.tight_layout()

# Sauvegarde
plt.savefig('images/question_01_adhésions_avec_valeurs.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

# 2. relation des adhésions

```python
df_relation = pandas.read_sql("""
SELECT pe.pk_personne, pe.nom AS nom,  pe.prenom AS prenom,  pe.genre AS genre,  pe.adhesion AS adhesion, 
    tr.definition AS presente_par,  p2.nom AS nom_P2, p2.prenom AS prenom_P2
FROM personne AS pe
LEFT JOIN relation AS r 
    ON r.fk_personne1 = pe.pk_personne
LEFT JOIN type_R AS tr 
    ON tr.pk_type_r = r.fk_type_r
LEFT JOIN personne AS p2 
    ON p2.pk_personne = r.fk_personne2
WHERE tr.definition = 'présenté_par'
ORDER BY pe.adhesion ASC;
""", connection)

df_relation.to_csv('CSV/question_01_relations_adhésion.csv')

df_relation.head(1000)
```

```python
df_nombres_relation = pandas.read_sql("""
SELECT 
    p2.pk_personne AS pk_presentateur,
    p2.nom AS nom_presentateur,
    p2.prenom AS prenom_presentateur,
    COUNT(pe.pk_personne) AS nb_personnes_presentees
FROM relation AS r
LEFT JOIN type_R AS tr 
    ON tr.pk_type_r = r.fk_type_r
LEFT JOIN personne AS pe 
    ON pe.pk_personne = r.fk_personne1
LEFT JOIN personne AS p2 
    ON p2.pk_personne = r.fk_personne2
WHERE tr.definition = 'présenté_par'
GROUP BY p2.pk_personne, p2.nom, p2.prenom
ORDER BY nb_personnes_presentees DESC;
""", connection)

df_nombres_relation.head(1000)
```
