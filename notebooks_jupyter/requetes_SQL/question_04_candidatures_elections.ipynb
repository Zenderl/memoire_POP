# 4. élections législatives

```python
import sqlite3
import pandas
import matplotlib.pyplot as plt
import seaborn as sns


connection = sqlite3.connect('C:/Users/LZender/Documents/Travail/unine/histoire/mémoire/AEN/grand_conseil/memoire_POP/liste_grand_conseil.db')


cursor = connection.cursor()
```

```python
annees = list(range(1961, 1982,5))

periodes = []

for index, annee_item in enumerate(annees):
    if index + 1 < len(annees):
        periodes.append({
            'start': annees[index],
            'end': annees[index + 1] - 1
        })

for periode_item in periodes:
    start = periode_item['start']
    end = periode_item['end']
    cursor.execute("""
SELECT COUNT(pk_evenement) FROM personne as pe
LEFT JOIN participation as pa ON pa.fk_personne = pe.pk_personne
LEFT JOIN evenement as e ON e.pk_evenement = pa.fk_evenement
WHERE genre = 'F' AND pa.annee BETWEEN ? AND ?
AND e.definition ='legislatif'
""", [start, end])
    result_femmes = cursor.fetchall()[0][0]
    cursor.execute("""
SELECT COUNT(pk_evenement) FROM personne as pe
LEFT JOIN participation as pa ON pa.fk_personne = pe.pk_personne
LEFT JOIN evenement as e ON e.pk_evenement = pa.fk_evenement
WHERE genre = 'M' AND pa.annee BETWEEN ? AND ?
AND e.definition ='legislatif'
""", [start, end])
    result_hommes = cursor.fetchall()[0][0]
    periode_item['femmes'] = result_femmes
    periode_item['hommes'] = result_hommes


print(periodes)
```

```python
df_candidatures_elections = pandas.DataFrame.from_dict(periodes)

df_candidatures_elections['total'] = df_candidatures_elections['femmes'] + df_candidatures_elections['hommes']

df_candidatures_elections['pourcentage_femmes'] = df_candidatures_elections['femmes'] / df_candidatures_elections['total'] * 100.00
df_candidatures_elections['pourcentage_hommes'] = df_candidatures_elections['hommes'] / df_candidatures_elections['total'] * 100.00

df_candidatures_elections.head()
```

```python
df_copy_candidatures_elections = df_candidatures_elections.drop(columns=['femmes', 'hommes', 'total'])

df_copy_candidatures_elections.head()

df_copy_candidatures_elections.to_csv('CSV/question_04_candidatures_elections.csv')
```

### répartitions par périodes en graphe

```python
df_modif_candidatures_elections = df_copy_candidatures_elections.copy()
df_modif_candidatures_elections['periodes'] = df_modif_candidatures_elections['start'].astype(str) + ' ' + df_modif_candidatures_elections['end'].astype(str)
df_modif_candidatures_elections = df_modif_candidatures_elections.drop(columns=['start', 'end'])


df_modif_candidatures_elections.head()
```

```python
df_long_elections = df_modif_candidatures_elections.melt(
    id_vars='periodes',
    value_vars=['pourcentage_femmes', 'pourcentage_hommes'],
    var_name='genre',
    value_name='pourcentage'
)

df_long_elections.head()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df_long_elections, x='periodes', y='pourcentage', hue='genre', palette='pastel')

# Rotation des labels x
plt.xticks(rotation=45, ha='right')

# Titre et axes
plt.title("Répartition des candidatures par genre  (en % sur l'ensemble)")
plt.ylabel("Pourcentage")
plt.xlabel("Période")

# Ajouter les valeurs sur les barres
for p in ax.patches:
    hauteur = p.get_height()
    if not pd.isna(hauteur):  # au cas où il y a des NaN
        ax.annotate(f'{hauteur:.1f}%', 
                    (p.get_x() + p.get_width() / 2., hauteur),
                    ha='center', va='bottom', fontsize=9)
        
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Ajustement des marges
plt.tight_layout()

# Sauvegarde
plt.savefig('images/question_04_candidatures_élections_avec_valeurs.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
