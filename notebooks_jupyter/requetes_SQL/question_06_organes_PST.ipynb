```python
import sqlite3
import pandas 
import matplotlib.pyplot as plt
import seaborn as sns
connection = sqlite3.connect('C:/Users/LZender/Documents/Travail/unine/histoire/mémoire/AEN/grand_conseil/memoire_POP/liste_grand_conseil.db')


cursor = connection.cursor()
```

## 6.1 participation nominative comme délégué.e.s

```python
df_delegues_nom = pandas.read_sql("""
SELECT pe.nom as nom, pe.prenom as prenom, pe.genre as genre, f.nom as delegue, e.annee_debut as debut, e.annee_fin as fin FROM personne as pe
LEFT JOIN exercice as e ON e.fk_personne = pe.pk_personne
LEFT JOIN fonction as f ON f.pk_fonction = e.fk_fonction

where f.pk_fonction is not null AND f.nom ='delegue_pst'
group by pk_personne
order by debut ASC
""", connection)

df_delegues_nom.to_csv('CSV/question_06_noms_delegues.csv')

df_delegues_nom.head(1000)
```

### 6.1.1 noms des déléguéEs

```python
df_delegues_genre_F = pandas.read_sql("""
SELECT pe.nom as nom, pe.prenom as prenom, pe.genre as genre, e.annee_debut as debut, e.annee_fin as fin FROM personne as pe
LEFT JOIN exercice as e ON e.fk_personne = pe.pk_personne
LEFT JOIN fonction as f ON f.pk_fonction = e.fk_fonction

where f.pk_fonction is not null AND f.nom ='delegue_pst' AND pe.genre='F'
group by pk_personne
order by debut ASC
""", connection)

df_delegues_genre_F.head(1000)
```

a noter que moins de 1/4 des délégué.e.s du POPNe étaient des femmes et qu'il faut attendre 1968 pour qu'il y ait une déléguéE

## 6.2 participation au comité central et/ou au bureau politique

```python
df_CC_BP = pandas.read_sql("""
SELECT pe.nom as nom, pe.prenom as prenom, pe.genre as genre, f.nom as poste, e.annee_debut as debut, e.annee_fin as fin FROM personne as pe
LEFT JOIN exercice as e ON e.fk_personne = pe.pk_personne
LEFT JOIN fonction as f ON f.pk_fonction = e.fk_fonction

where f.pk_fonction is not null AND f.nom ='comite_central' OR f.nom ='bureau_politique'
order by debut ASC
""", connection)

df_CC_BP.to_csv('CSV/question_06_noms_comite_central_bureau_politique.csv')

df_CC_BP.head(1000)
```
